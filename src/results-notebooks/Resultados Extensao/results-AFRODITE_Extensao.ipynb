#### Processamento dos arquivos do servidor para obtenção do tempo de treinamento

```python
m1_result_server = open("../../resultados_experimentos/balanceado_4clientes/sampling-out_servidor.txt", 'r')
m2_result_server = open("../../resultados_experimentos/balanceado_4clientes/full-out_servidor.txt", 'r')
m3_result_server = open("../../resultados_experimentos/desbalanceado_4clientes/sampling-out_servidor.txt", 'r')
m4_result_server = open("../../resultados_experimentos/desbalanceado_4clientes/full-out_servidor.txt", 'r')

m1_lines = m1_result_server.readlines()
m2_lines = m2_result_server.readlines()
m3_lines = m3_result_server.readlines()
m4_lines = m4_result_server.readlines()

files = [m1_lines, m2_lines, m3_lines, m4_lines]
```

```python
times = [[],[],[],[]]
days = [[],[],[],[]]

for client, f in enumerate(files):
  print(client)
  for idx, line in enumerate(f):
    if (line != '\n') and ('evaluate_round' in line) and ('received' in line):
      times[client].append( line.split(' ')[3].split(',')[0])
      days[client].append(line.split(' ')[2])
```

```python
seconds = [[],[],[],[]]
hours = [[],[],[],[]]

print('Qtd. Variações:',len(times),'\n','Qtd. medidas:', len(times[0]))
for i in range(0,len(times)):
  for j in range(0, len(times[0])):

    AcDepart = days[i][j] + ' ' + times[i][j]
    AcDepart_ = datetime.strptime(AcDepart, '%Y-%m-%d %H:%M:%S')
    AcArrival = days[i][0] + ' ' + times[i][0]
    AcArrival_ = datetime.strptime(AcArrival, '%Y-%m-%d %H:%M:%S')
    ActualSeconds = (AcDepart_ - AcArrival_).total_seconds()
    ActualHours = ActualSeconds/3600
    seconds[i].append(ActualSeconds)
    hours[i].append(ActualHours)
```

```python
for i in range(len(times)):
  print(seconds[i][-1])
```

```python
x = [i for i in range(0, len(seconds[0]))]
labels=['M1: RS','M2', 'M3: RS', 'M4' ]
cor = ['b', 'g', 'r', 'm']

plt.figure(figsize=(10,8))
for i in range(0, len(seconds)):
  plt.plot(x, seconds[i], color=cor[i],label=labels[i])
plt.ylabel('Tempo (s)', fontsize=10)
plt.title('Tempo de treinamento')
plt.xlabel('Época', fontsize=10)
plt.legend()
plt.show()

```
